##### Метрики:
Самой очевидной метрикой для расчёта пользователей покинувших игру является Churm Rate. В идеале данный показатель должен стремиться к нулю: 


\begin{equation}
ChurmRate = \frac{Количество \, ушедших\, клиентов}{Общее \, количество \, клиентов}
\end{equation}

Хорошей метрикой для расчёта удержания клиентов является Customer Retention Rate. Чем выше данный показатель, тем лучше:


\begin{equation}
CustomerRetentionRate = \frac{Количество\,клиентов\,на\,конец\,периода - Количество\,новых\,клиентов\,за\,период}{Количество\,клиентов\,на\,начало\,периода}
\end{equation}

Хорошей метрикой также является процент пользователей зашедших в приложение на N-й день после установки приложения:


\begin{equation}
Retention(N) = \frac{Кол-во\,пользователей,\,зашедших\,в\,день\,N}{Кол-во\,пользователей,\,установивших\,приложение\,N дней\,назад}*100 \%
\end{equation}


Расчитать данные показатели можно, к примеру, при помощи Статистики Steam.

Первая игровая сессия очень сильно влияет на то, останется игрок или уйдёт, и в идеале она должна произвести максимально приятное впечатление на игрока. Для этого игре необходим интуитивный туториал и демонстрация в нём всех основных "фишек".

Помимо описанных выше характеристик следует также вести мониторинг активности на сторонних сервисах (например, стримы на Twitch). Уменьшение людей смотрящих стримы и самих стримеров показывает нисходящий тренд активных игроков. 

##### Что можно считать моментом покидания пользователем игры? 

Во-первых важно установить порог времени бездействия игрока, по прошествии которого можно считать игрока ушёдшим. Данное значение подбирается индивидуально для каждого приложения и сильно влияет на итоговую оценку. Крайне важно балансировать между быстротой реагирования данной метрики и её стабильностью: недавно ушёдший игрок может вернуться вновь, а слишком долгое время бездействия даст очень запоздалую и от того малополезную оценку. 

Во-вторых: если игра подразумевает наличие микротранзакций, ушедшим игроком монжо считать того, кто перестал их вносить, либо отказался от платной подписки, дающей преференции.

В-третьих: важно смотреть на количество зашедших в игру игроков, после крупного обновления или введения в игру новых механик. Вернувшийся игрок, скорее всего, сохраняет интерес к игре, но в силу личных причин её на время покинул (например игра просто приелась из-за отсутствия нового контента). В качестве примера можно привести многострадальную Quake Champions: после введения крупных обновлений онлайн игры многократно возрастал, что указывает на то, что люди сохраняли к ней интерес. Потом, правда, снова падал до прежних показателей, потому что комьюнити игры эти обновления пришлись не по вкусу.


##### Какие меры можно предпринять для удержания пользователей в игре?
Если техническое состояние игры хорошее, но пользователи продолжают её покидать можно предложить несколько вариантов по их удержанию. Если же техническое состояние игры остаётся стабильно плохим, игроков вряд ли что-то в ней удержит и в первую очередь необходимо улучшать его.

1) При прочьих равных самым главным условием, особенно в соревновательных играх по типу FPS, является баланс сложности. Даже если игра идеальна в других аспектах, дисбаланс напрочь убивает интерес игроков к продолжению. К примеру если в одну сессию закидывают ветеранов игры и относительно новых игроков, последние будут постоянно проигрывать и уйдут из игры. Напротив, при низком уровне игроки заскучают и также уйдут. Рост уровня сложности должен проходить постепенно.

2) Как уже было сказано ранее, постоянное введение обновлений, новых механик и режимов игры может не только удержать пользователей в игре, но и привлечь новых. Это может также оказать и чисто психологический эффект: часть игроков будет видеть, что разработчики усердно работают над её качеством и продолжат играть, даже если она им наскучила, надеясь что грядущие обновления вернут интерес. К этому делу, правда, надо подходить осторожно, главый принцип - "не сломать то, что и так хорошо работает".   

3) Анализ отзывов об игре: никто не скажет лучше что в игре плохо или хорошо, чем сами игроки. 

4) Необходимо также понимать кто является целевой аудиторией данной игры и при построении стратегии учитывать данный факт.

В целом методику выявляения оттока пользователей можно свести к расчёту основных показателей и дальнейшему анализу отзывов об игре. При данном анализе также следует различать конструктивную критику и ревью-бомбинг (ситуация, при которой многочисленные группы людей оставляют отрицательные пользовательские отзывы) который может являться результатом критики как за низкое качество продукта, так и по не связанным с этим причинами, которые носят политический или культурный характер, что также необходимо учитывать.
